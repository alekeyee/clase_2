***5. Reescriba el siguiente programa reemplazando la lista de comprensión con un while-loop.***


```python
list_1 = [1, 2, 3, 4]
result  =  [digit ** 2 for digit in list_1 if digit % 2 == 0]
print(result)
```

```python
list_1 = [1, 2, 3, 4]
while list_1 :
    result  =  [digit ** 2 for digit in list_1 if digit % 2 == 0]
    print(result)
    break
```
